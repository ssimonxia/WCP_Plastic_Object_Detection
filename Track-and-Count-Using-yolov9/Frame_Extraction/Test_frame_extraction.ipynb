```python
import cv2
import numpy as np
import random
from os import makedirs
from os.path import exists
import os
```

```python
save_path = "extracted_frame_large_2"
if not exists(save_path):
    makedirs(save_path)
    
```

```python
p = os.listdir()
p
video_path = []
for i in p:
    if i.split(".")[-1] == "MP4":
        video_path.append(i)
video_path = ["GX010093.MP4"]
video_path
```

```python
frame_count = 0
for path in video_path:
    cap = cv2.VideoCapture(path)
    num_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    if (cap.isOpened() == False):
        print("Error opening video stream or file")
    print(num_frames)
    flag = 0
    while (cap.isOpened()):
        ret, frame = cap.read()
        num = random.randint(0, num_frames-1)
        print(num, flag)
        if num >= 0 and num <= 100:
            if ret == True:
                cv2.imwrite(f"{save_path}/{frame_count}.png", frame)
                frame_count += 1
        flag+=1
        if flag >= num_frames:
            break
    cap.release()
```
